```python
#data uit mijn bestand naar csv zetten
#daarna een plot maken van de achtergrondruis en dus simpelweg mijn data
zonder_leakage = 'spectrum_zonder_leakage'
zonder_leakage_csv = 'spectrum_zonder_leakage.csv'

with open(zonder_leakage, 'r') as infile, open(zonder_leakage_csv, 'w') as outfile:
    for line in infile:
        values = line.strip().split()
        outfile.write(','.join(values) + '\n')

```

```python
import pandas as pd

df = pd.read_csv('spectrum_zonder_leakage.csv', sep=',', header=None)
df = df.iloc[1:].reset_index(drop=True)
df = df.apply(pd.to_numeric, errors='coerce') #alle waarden numeriek zetten want fout meldingen drive me crazy
```

```python

pulseheight_cesium = df.iloc[:, 0].values  
counts_cesium = df.iloc[:, 1].values      

print(pulseheight_cesium[:10])
print(counts_cesium[:10]) #puur om te checken
```

```python

import matplotlib.pyplot as plt

plt.xlim(0, 4000)
plt.plot(pulseheight_cesium, counts_cesium)
plt.title("Achtergrondruis bij Cesium-137")
plt.xlabel("Pulshoogte (mV)")
plt.ylabel("Counts")
plt.axvline(2085, color='grey', linestyle='--', alpha = 0.4)
plt.axvline(2250, color='grey', linestyle='--', alpha = 0.4)
plt.axvline(2168.48, color='black', linestyle='--', label="Fotopiek", alpha = 0.7)

plt.show()
```

```python
#van Compton scattering data een plot maken maar eerst csv file van maken
#hetzelfde methode als hierboven alleen in een code 

import pandas as pd
import matplotlib.pyplot as plt

compton_data = "theta1_6dec_spectrum3"
compton_data_csv = "theta1_6dec_spectrum3.csv"

with open(compton_data, 'r') as infile, open(compton_data_csv, 'w') as outfile:
    for line in infile:
        values = line.strip().split()
        outfile.write(','.join(values) + '\n')


df_compton = pd.read_csv(compton_data_csv, sep=',', header=None)
df_compton = df_compton.iloc[1:].reset_index(drop=True)
df_compton = df_compton.apply(pd.to_numeric, errors='coerce') #alle waarden numeriek zetten want fout meldingen drive me crazy

pulseheight_cesium_compton = df_compton.iloc[:, 0].values  
counts_cesium_compton = df_compton.iloc[:, 1].values      

plt.xlim(0, 4000)
plt.plot(pulseheight_cesium_compton, counts_cesium_compton)
plt.title("Comptonverstrooiing bij Cesium-137")
plt.ylabel("Counts")
plt.xlabel("Pulshoogte (mV)")
plt.axvline(2085, color='grey', linestyle='--', alpha = 0.4)
plt.axvline(2250, color='grey', linestyle='--', alpha = 0.4)
plt.axvline(2168.48, color='black', linestyle='--', label="Fotopiek", alpha = 0.7)
plt.show()
```
